**SMA**

In this exercise, we'll learn abuot how to calculate <font color=blue>Stock SMA, simple moving average</font>. There are other methods out there for calculating SMA, please google.

```python
f = open('FB.csv')
f.seek(0)                 # optional, move the 'pointer' to beginning of file
all_data = f.read()
# all_data
```

```python
all_data_list = all_data.split('\n')

# for the scope of this practice, we don't really need headers. 
# It's nice to know how to calculate headers for later usage
headers = all_data_list[0].split(',')
headers
```

```python
# peek into the data
all_data_list[:10]
```

```python
data = [x.split(',') for x in all_data_list[1:] if x != '']
dates = [d[0] for d in data]             #  list comprehension
prices = [float(d[5]) for d in data]     #  convert string price to float price
# print(dates[:10])
# len(prices)
# prices
```

```python
# here, the n-th day's SMA is defined as the average price of the 8 days before the n-th day, not including itself. 
# so later we'll see in the plot, in the beginning we're missing some SMA data, that's due to the way we do calculation.
def sma(prices, nday):
    sma_data = [0] * len(prices)
    for i in range(nday, len(prices)):
        sma_data[i] = sum(prices[i-nday:i]) / nday
    return sma_data
```

```python
# we're calculating sma8 (8 day moving averatma35 for this practice
period1, period2,  = 8, 35
data_sma1, data_sma2 = sma(prices, period1), sma(prices, period2)
labels = ['FB_SMA_%s' % x for x in (period1, period2)]
# print('Labels: ', labels)
# print(data_sma1[:10])
# print(data_sma2[:10])
```

```python
# 'nan': not a number. usually 'nan' is used for extremely large number, extremely small number, or missing data
def zero_to_nan(values):
    """Replace every 0 with 'nan' and return a copy."""
    return [float('nan') if x==0 else x for x in values]       # list comprehension, with if/else

# matplotlib skips 'nan' when plotting, this is a good way to 'drop' some invalid data
data_sma1 = zero_to_nan(data_sma1)
data_sma2 = zero_to_nan(data_sma2)
# data_sma2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,5))                # 14 inches wide, 5 inches height
plt.title('Facebook Stock 1 Year Data and SMA')
plt.plot(prices, linestyle='--', color='red', linewidth=2.0, label='FB') 
plt.plot(data_sma1, color='green', linestyle='-.', linewidth=3.0, label=labels[0])     
plt.plot(data_sma2, 'bo', markersize=1.2, label=labels[1])                    # bo: blue dots
plt.plot(dates, linestyle='-', color='pink', linewidth=2.0, label='date')                        # set y limit(range)
plt.legend()                              # show legend of 'FB' and 'FB_SMA3'
plt.savefig('facebook_sma8_sma35.png')    # 'savefig' needs to happen before 'plt.show', move it below and check the image saved
plt.show()
```
